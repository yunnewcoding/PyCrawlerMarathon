### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
# 先試著找出檔案中包含哪些縣市
with open('C:\\Users\\yun10\\Desktop\\爬蟲馬拉松\\Day003_example\\TAIWAN_72hr_CH.xml', 'r', newline='', encoding='utf-8') as f:

    import xmltodict
    doc=xmltodict.parse(f.read())
    data=doc["cwbopendata"]["dataset"]["locations"]["location"]  # 用記事本打開 xml 找標籤會好找很多
    
    # 由於在 xml 中相同的標籤，在 xmltodict 轉換時會被轉成一個 list，例如 <location>標籤下的資料都會被收在一個 list 裡，所以在取值時要加上[...]的格式
    # 接下來一個一個印出資料中的縣市
    
    for i in range(len(data)):            # 從下一行倒著想回來，data裡共有 22 組資料，從 1 取到第 i 個縣市，所以用 for 一個一個取出來
        print(data[i]["locationName"])    # 取出 locationName 標籤下的第 i 個值

print("==============我是分隔線==================")        
        
# 根據問答區，高雄的檔案在 64_72hr_CH.xml 
 
with open('C:\\Users\\yun10\\Desktop\\爬蟲馬拉松\\Day003_example\\64_72hr_CH.xml ', 'r', newline='', encoding='utf-8') as k:
        doc=xmltodict.parse(k.read())
        area=doc["cwbopendata"]["dataset"]["locations"]["location"]
        for i in range(len(area)):
            print(area[i]["locationName"])
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
with open('C:\\Users\\yun10\\Desktop\\爬蟲馬拉松\\Day003_example\\64_72hr_CH.xml ', 'r', newline='', encoding='utf-8') as k:
    import xmltodict
    doc=xmltodict.parse(k.read())
    data=doc["cwbopendata"]["dataset"]["locations"]["location"][0]["weatherElement"]
    print(data)
```
